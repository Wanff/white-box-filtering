## Language plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Data
languages = ['Dutch', 'Hungarian', 'Slovenian']
llamaguard_accs = [0.52, 0.56, 0.64]
transfer_probe_accs = [0.55, 0.55, 0.52]
recalibrated_probe_accs = [0.85, 0.93, 0.92]
ceiling_probe_accs = [1.0, 0.96, 0.94]

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

models = ['LlamaGuard', 'Transfer Probe', 'Recalibrated Probe', 'Ceiling Probe']

for i, language in enumerate(languages):
    accuracies = [llamaguard_accs[i], transfer_probe_accs[i], recalibrated_probe_accs[i], ceiling_probe_accs[i]]
    
    # Plot each model's accuracy as a separate bar
    for idx, model in enumerate(models):
        axs[i].bar(model, accuracies[idx], label=model if i == 0 else "")

    axs[i].set_ylim(0, 1.1)  # Extend y-axis slightly above 1 for visual comfort
    axs[i].set_ylabel('Accuracy')
    axs[i].set_xlabel(language)

    # Only add x-ticks and legend to the first subplot to avoid repetition
    if i == 0:
        axs[i].legend(title='Model')

    # Remove x-tick labels for all
    axs[i].set_xticks([])

plt.suptitle('Transfer accuracy across languages')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style
sns.set(style="whitegrid")

# Data
languages = ['Dutch', 'Hungarian', 'Slovenian']
llamaguard_accs = [0.52, 0.56, 0.64]
transfer_probe_accs = [0.55, 0.55, 0.52]
recalibrated_probe_accs = [0.85, 0.93, 0.92]
ceiling_probe_accs = [1.0, 0.96, 0.94]

# Creating DataFrame
data = {
    'Language': languages * 4,
    'Accuracy': llamaguard_accs + transfer_probe_accs + recalibrated_probe_accs + ceiling_probe_accs,
    'Model': ['LlamaGuard'] * 3 + ['Transfer Probe'] * 3 + ['Recalibrated Probe'] * 3 + ['Ceiling Probe'] * 3
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Language', y='Accuracy', hue='Model', data=df, palette='Reds')

# Adding labels and title
bar_plot.set_xlabel('', fontsize=12)
bar_plot.set_ylabel('Accuracy', fontsize=12)
bar_plot.set_title('Transfer accuracy across languages', fontsize=14)
plt.legend(title='Model')

# Show the plot
plt.show()
```
